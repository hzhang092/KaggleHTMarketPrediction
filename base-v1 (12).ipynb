```python
def categorize_features(df):
    """Categorize features based on their prefixes"""
    
    print("\nğŸ·ï¸ FEATURE CATEGORIZATION")
    print("="*50)
    
    feature_categories = {}
    
    # Define feature categories based on prefixes
    categories = {
        'Market_Dynamics': 'M',
        'Macro_Economic': 'E',
        'Interest_Rate': 'I',
        'Price_Valuation': 'P',
        'Volatility': 'V',
        'Sentiment': 'S',
        'Momentum': 'MOM',
        'Dummy_Binary': 'D'
    }
    
    # Categorize features
    for category, prefix in categories.items():
        if prefix == 'MOM':
            feature_categories[category] = [col for col in df.columns if col.startswith(prefix)]
        else:
            feature_categories[category] = [col for col in df.columns if col.startswith(prefix) and not col.startswith('MOM')]
    
    # Special columns
    special_cols = ['date_id', 'forward_returns', 'risk_free_rate', 'market_forward_excess_returns']
    feature_categories['Special'] = [col for col in special_cols if col in df.columns]
    
    # Display categorization
    total_features = 0
    for category, features in feature_categories.items():
        print(f"ğŸ“Š {category}: {len(features)} features")
        total_features += len(features)
    
    print(f"\nğŸ“ˆ Total categorized features: {total_features}")
    print(f"ğŸ“ˆ Total columns in dataset: {df.shape[1]}")
    
    return feature_categories

feature_categories = categorize_features(train_df)
```

```python
def analyze_feature_importance(df, feature_categories, max_features=100):
    """Analyze feature importance using various methods"""
    
    print("\nâ­ FEATURE IMPORTANCE ANALYSIS")
    print("="*50)
    
    if 'forward_returns' not in df.columns:
        print("âŒ No target variable found for feature importance analysis")
        return
    
    # Prepare data - remove non-feature columns
    feature_cols = []
    for category, features in feature_categories.items():
        if category != 'Special':
            feature_cols.extend(features)
    
    # Limit features for computational efficiency
    if len(feature_cols) > max_features:
        feature_cols = feature_cols[:max_features]
        print(f"âš ï¸ Limited to first {max_features} features for computational efficiency")
    
    # Create feature matrix
    X = df[feature_cols].fillna(0)
    y = df['forward_returns'].fillna(0)
    
    # Remove rows where target is missing
    mask = ~df['forward_returns'].isna()
    X = X[mask]
    y = y[mask]
    if len(X) == 0:
        print("âŒ No valid data for feature importance analysis")
        return
    
    print(f"ğŸ“Š Analyzing {len(feature_cols)} features with {len(X)} samples")
    
    # 1. Correlation-based importance
    corr_importance = abs(X.corrwith(y)).sort_values(ascending=False)
    
    # 2. Mutual Information
    try:
        mi_scores = mutual_info_regression(X, y, random_state=42)
        mi_importance = pd.Series(mi_scores, index=feature_cols).sort_values(ascending=False)
    except:
        print("âš ï¸ Mutual information calculation failed")
        mi_importance = pd.Series(index=feature_cols)
    
    # Combine results
    importance_df = pd.DataFrame({
        'Correlation': corr_importance,
        'Mutual_Info': mi_importance
    }).fillna(0)
    
    # Create composite score
    importance_df['Composite_Score'] = (
        importance_df['Correlation'].rank(pct=True) + 
        importance_df['Mutual_Info'].rank(pct=True)
    ) / 2
    
    importance_df = importance_df.sort_values('Composite_Score', ascending=False)
    
    print("\nğŸ† TOP 50 MOST IMPORTANT FEATURES:")
    display(importance_df.head(10))
    
    return importance_df

importance_results = analyze_feature_importance(train_df, feature_categories)
```

```python
def ad_sharpe_ratio_scorer(y_pred, y):
    
    y_pred = y_pred.astype(float)
    y_true = y.astype(float)
    
    # 3. è®¡ç®—ç­–ç•¥çš„æ¯æ—¥æ”¶ç›Š
    # ç­–ç•¥é€»è¾‘ï¼šé¢„æµ‹ä¸ºæ­£æ—¶ä¹°å…¥ï¼Œé¢„æµ‹ä¸ºè´Ÿæ—¶ä¸æ“ä½œ
    positions = np.where(y_pred > 0, y_pred, 0)
    #positions = np.clip(positions, 0, 2)
    strategy_returns = positions * y_true
    
    # 4. è®¡ç®—å¤æ™®æ¯”ç‡
    mean_return = np.mean(strategy_returns)
    std_return = np.std(strategy_returns)
    
    # å¢åŠ ä¸€ä¸ªå°çš„ epsilon ä»¥é¿å…é™¤ä»¥é›¶
    epsilon = 1e-8
    sharpe_ratio = mean_return / (std_return + epsilon)
    
    # 5. è¿”å›å¤æ™®æ¯”ç‡ã€‚scikit-learn ä¼šå°è¯•æœ€å¤§åŒ–è¿™ä¸ªå€¼ã€‚
    return sharpe_ratio
```

```python
class DataFrameWrapper(BaseEstimator, TransformerMixin):
    """
    ä¸€ä¸ªåŒ…è£…å™¨ï¼Œç”¨äºåŒ…è£¹ scikit-learn çš„è½¬æ¢å™¨ (å¦‚ SimpleImputer)ï¼Œ
    ä»¥ç¡®ä¿å…¶è¾“å‡º *å§‹ç»ˆ* æ˜¯ä¸€ä¸ª Pandas DataFrameã€‚
    """
    def __init__(self, transformer):
        self.transformer = transformer
        
    def fit(self, X, y=None):
        """
        åœ¨ fit æœŸé—´ï¼Œå­¦ä¹  X_train çš„åˆ—åï¼Œ
        å¹¶æ‹Ÿåˆå†…éƒ¨çš„ transformerã€‚
        """
        # (å‡è®¾ X æ˜¯ä¸€ä¸ª DataFrame)
        self.columns_ = X.columns
        self.index_ = X.index
        
        self.transformer.fit(X, y)
        return self

    def transform(self, X):
        """
        åœ¨ transform æœŸé—´ï¼Œè°ƒç”¨å†…éƒ¨çš„ transformerï¼Œ
        ç„¶åç«‹å³å°† NumPy è¾“å‡ºé‡å»ºå› DataFrameã€‚
        """
        # X å¯èƒ½æ˜¯ X_train æˆ– X_test
        
        # 1. å†…éƒ¨ transformer (ä¾‹å¦‚ SimpleImputer) è¾“å‡º NumPy
        numpy_output = self.transformer.transform(X)
        
        # 2. ç«‹å³é‡å»º DataFrame
        #    ä½¿ç”¨åœ¨ .fit() æœŸé—´å­¦åˆ°çš„åˆ—å
        #    ä½¿ç”¨ .transform(X) ä¸­ä¼ å…¥çš„ X çš„ç´¢å¼•
        df_output = pd.DataFrame(
            numpy_output,
            columns=self.columns_,
            index=X.index
        )
        
        return df_output
```

```python

import pandas as pd
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.linear_model import LinearRegression
from scipy.signal import detrend

class TopKFourierFeatures(BaseEstimator, TransformerMixin):
    """
    æœ€ç»ˆç‰ˆ:
    1. Fit: ä½¿ç”¨FFTåœ¨y_trainä¸Šå‘ç°ä¿¡å·æœ€å¼ºçš„kä¸ªå‘¨æœŸï¼Œå¹¶ç‹¬ç«‹æ‹Ÿåˆkä¸ªsin+cosæ¨¡å‹ã€‚
    2. Transform: å°†kä¸ªæ¨¡å‹çš„é¢„æµ‹å€¼åŠ æ€»ï¼Œç”Ÿæˆä¸€ä¸ªå•ä¸€çš„'fourier_top_k_sum'ç‰¹å¾ã€‚
    """
    def __init__(self, k=20, date_col='date_id',max_period=252):
        self.k = k
        self.date_col = date_col
        self.max_period = max_period
        # æ³¨æ„: åœ¨è¿™ä¸ªç‰ˆæœ¬ä¸­ï¼Œæˆ‘ä»¬å°†ä¸å†æŠŠçº¿æ€§è¶‹åŠ¿ä½œä¸ºå•ç‹¬çš„è¾“å‡ºç‰¹å¾ï¼Œ
        # å› ä¸ºç›®æ ‡æ˜¯åªè¾“å‡ºä¸€ä¸ªæ€»å‘¨æœŸç‰¹å¾ã€‚ä½†è¶‹åŠ¿æ¨¡å‹ä»ç„¶åœ¨å†…éƒ¨ç”¨äºå»è¶‹åŠ¿ã€‚
        self.trend_model_ = None 
        self.cycle_models_ = []

    def fit(self, X, y):
        if self.date_col not in X.columns:
            raise ValueError(f"Input DataFrame X must have a '{self.date_col}' column.")

        y_fit = pd.Series(y, index=X.index).dropna()
        X_fit = X.loc[y_fit.index]
        time_index = X_fit[self.date_col].values
        
        # 1. æ‹Ÿåˆæ•´ä½“çº¿æ€§è¶‹åŠ¿å¹¶å»è¶‹åŠ¿ (ç”¨äºåç»­å‘¨æœŸæ‹Ÿåˆ)
        self.trend_model_ = LinearRegression()
        self.trend_model_.fit(time_index.reshape(-1, 1), y_fit)
        detrended_series = y_fit - self.trend_model_.predict(time_index.reshape(-1, 1))

        # 2. ä½¿ç”¨FFTå‘ç°Top-Kä¸ªæœ€å¼ºçš„å‘¨æœŸ
        n_samples = len(detrended_series)
        if n_samples < 50: 
            return self
            
        fft_result = np.fft.fft(detrended_series)
        frequencies = np.fft.fftfreq(n_samples, d=1)
        n_positive_freqs = n_samples // 2
        amplitudes = np.abs(fft_result[:n_positive_freqs])
        positive_frequencies = frequencies[:n_positive_freqs]
        
        with np.errstate(divide='ignore'):
            periods = np.where(positive_frequencies != 0, 1 / positive_frequencies, np.inf)
        
        valid_mask = (periods >= 2) & (periods != np.inf)
        if self.max_period is not None and self.max_period > 0:
            valid_mask &= (periods <= self.max_period)
            
        periods_filtered, amplitudes_filtered = periods[valid_mask], amplitudes[valid_mask]
        
        top_k_indices = np.argsort(amplitudes_filtered)[::-1][:self.k]
        top_k_periods = periods_filtered[top_k_indices]

        # 3. å¯¹æ¯ä¸ªå‘ç°çš„å‘¨æœŸï¼Œç‹¬ç«‹æ‹Ÿåˆä¸€ä¸ªsin+cosæ¨¡å‹
        self.cycle_models_ = []
        for p in top_k_periods:
            sin_feature = np.sin(2 * np.pi * time_index / p)
            cos_feature = np.cos(2 * np.pi * time_index / p)
            X_cycle = pd.DataFrame({'sin': sin_feature, 'cos': cos_feature})
            
            cycle_model = LinearRegression()
            # æ‰€æœ‰æ¨¡å‹éƒ½æ‹ŸåˆåŒä¸€ä¸ªå»è¶‹åŠ¿åçš„åºåˆ—
            cycle_model.fit(X_cycle, detrended_series)
            self.cycle_models_.append({'period': p, 'model': cycle_model})
        return self

    def transform(self, X):
        if not self.cycle_models_:
            print("  - Warning: No cycles were learned. Returning original DataFrame.")
            return X
            
        X_transformed = X.copy()
        time_index = X[self.date_col].values
        
        # åˆå§‹åŒ–ä¸€ä¸ªå…¨é›¶æ•°ç»„ï¼Œç”¨äºç´¯åŠ æ‰€æœ‰å‘¨æœŸå‡½æ•°çš„è´¡çŒ®
        total_cycle_prediction = np.zeros(len(X))

        # éå†æ‰€æœ‰å­¦ä¹ åˆ°çš„å‘¨æœŸæ¨¡å‹
        for cycle_info in self.cycle_models_:
            p = cycle_info['period']
            model = cycle_info['model']
            
            # ä¸ºå½“å‰æ•°æ®(X)åˆ›å»ºsin/cosç‰¹å¾
            sin_feature = np.sin(2 * np.pi * time_index / p)
            cos_feature = np.cos(2 * np.pi * time_index / p)
            X_cycle_transform = pd.DataFrame({'sin': sin_feature, 'cos': cos_feature})
            
            # è®¡ç®—è¿™ä¸ªå‘¨æœŸå‡½æ•°çš„è´¡çŒ®å€¼
            cycle_component = model.predict(X_cycle_transform)
            
            # å°†è¿™ä¸ªå‘¨æœŸçš„è´¡çŒ®ç´¯åŠ åˆ°æ€»å’Œä¸­
            total_cycle_prediction += cycle_component

        # å°†æœ€ç»ˆçš„â€œæ€»å‘¨æœŸä¿¡å·â€ä½œä¸ºä¸€ä¸ªæ–°ç‰¹å¾æ·»åŠ è¿›å»
        X_transformed['fourier_top_k_sum'] = total_cycle_prediction
        
        return X_transformed
```

```python
class create_cluster(BaseEstimator, TransformerMixin):
    
    def __init__(self, feature_categories, K = 0, window=50):
        self.feature_categories = feature_categories
        self.window = window
        self.K = K 
        
    def _detrend_and_standardize(self, df):
        """
        Detrend and standardize features based on their categories.
        Price/Economic series: absolute change
        Indicators: rolling Z-score
        """
        df_transformed = df.copy()
        window = self.window # ä»ç±»å±æ€§è·å–çª—å£å¤§å°
        
        for category, features in self.feature_categories.items():
            # ç¡®ä¿åªå¤„ç†å­˜åœ¨çš„åˆ—
            features = [f for f in features if f in df_transformed.columns]
            
            if category in ['Price_Valuation', 'Macro_Economic']:
                # Absolute change for price/economic series
                for feature in features:
                    df_transformed[feature] = df_transformed[feature].diff().fillna(0)
                    
            elif category not in ['Dummy_Binary', 'Special']:
                # Rolling Z-score for indicators
                for feature in features:
                    rolling_mean = df_transformed[feature].rolling(window=window).mean()
                    rolling_std = df_transformed[feature].rolling(window=window).std()
                    
                    # ç”¨ 1.0 æ›¿æ¢ 0ï¼Œä»¥é˜²æ­¢é™¤ä»¥é›¶
                    rolling_std.replace(0.0, 1.0, inplace=True) 
                    
                    df_transformed[feature] = (df_transformed[feature] - rolling_mean) / rolling_std
                    
                    # å¡«å……æ»šåŠ¨çš„ Z-Score äº§ç”Ÿçš„ NaN
                    df_transformed[feature] = df_transformed[feature].fillna(0)
            
            # Dummy_Binary å’Œ Special ç±»åˆ«è¢«è‡ªåŠ¨è·³è¿‡
            
        return df_transformed
    # ----------------------------------------------------------------------

    def fit(self, X, y=None):
        # 1. å‡†å¤‡è¦èšç±»çš„ç‰¹å¾åˆ—
        feature_cols = []
        for category, features in self.feature_categories.items():
            if category != 'Special':
                feature_cols.extend(features)
        
        self.feature_cols = [f for f in feature_cols if f in X.columns]
        X_cluster = X[self.feature_cols]
        
        # 2. Imputer (ä»ç„¶éœ€è¦ï¼Œä»¥é˜²åŸå§‹æ•°æ®æœ‰ NaN)
        self.imputer = SimpleImputer(strategy='median').set_output(transform="pandas")
        X_imputed = self.imputer.fit_transform(X_cluster)
        
        # 3. åº”ç”¨é›†æˆçš„ detrend å’Œ standardize æ–¹æ³•
        X_processed = self._detrend_and_standardize(X_imputed)
        
        # 4. å¯»æ‰¾æœ€ä½³ K
        if  (self.K == 0):
            silhouette_scores = []
            k_range = range(2, 11)  # å›ºå®šçš„ã€å®‰å…¨çš„èŒƒå›´
            
            for k in k_range:
                kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
                cluster_labels = kmeans.fit_predict(X_processed.values)
                silhouette_avg = silhouette_score(X_processed.values, cluster_labels)
                silhouette_scores.append(silhouette_avg)
                self.K = k_range[np.argmax(silhouette_scores)]
                
        print(f"\nOptimal number of clusters (Detrended): {self.K}")
        
        # 5. è®­ç»ƒæœ€ç»ˆçš„ KMeans æ¨¡å‹
        self.kmeans = KMeans(n_clusters=self.K, random_state=42, n_init=10)
        self.kmeans.fit(X_processed.values)
        return self
        
    def transform(self, X):
        X_transformed = X.copy()
        
        # 1. ä»…é€‰æ‹©ç›¸å…³åˆ—
        X_cluster = X[self.feature_cols]
        
        # 2. åº”ç”¨ *å·²
        # è®­ç»ƒ* çš„ Imputer
        X_imputed = self.imputer.transform(X_cluster)
        
        # 3. åº”ç”¨ *ç›¸åŒ* çš„ detrend å’Œ standardize è½¬æ¢
        X_processed = self._detrend_and_standardize(X_imputed)
        
        # 4. åº”ç”¨ *å·²è®­ç»ƒ* çš„ KMeans
        cluster_labels = self.kmeans.predict(X_processed.values)
        
        X_transformed["cluster"] = cluster_labels
        return X_transformed
```

```python
log2 = True 

print(f"\nå¼€å§‹æ‰§è¡Œå•ä¸€æ—¶æ®µåˆ†æ...")
print(f"=== ç›®æ ‡: è®¡ç®— {new_model_descirption} çš„ Top 20 äº¤äº’ç»„åˆ ===")

n_samples = len(X)
train_size = int(n_samples * 0.8) 
X_train, y_train = X.iloc[2000:train_size], y.iloc[2000:train_size]
X_val, y_val = X.iloc[train_size:], y.iloc[train_size:]

print(f"è®­ç»ƒå‘¨æœŸ: index 2000 to {train_size-1} | æµ‹è¯•å‘¨æœŸ: index {train_size} to {n_samples-1}")


X_train_transformed2 = preprocessor_pipeline2.fit_transform(X_train, y_train)
model2.fit(X_train_transformed2, y_train)
X_val_transformed2 = preprocessor_pipeline2.transform(X_val)
preds_new = model2.predict(X_val_transformed2)
r2_val = r2_score(y_val, preds_new)
sharpe_val = ad_sharpe_ratio_scorer(preds_new, y_val) 
# ---

print(f"æ ·æœ¬å¤– RÂ² (R-squared): {r2_val:.4f}")
print(f"æ ·æœ¬å¤–å¤æ™®ç‡ (Sharpe Ratio): {sharpe_val:.4f}")

print(f"--- æ­£åœ¨è®¡ç®— SHAP äº¤äº’å€¼ (è¿™å¾ˆæ…¢)... ---")
try:
    explainer2 = shap.TreeExplainer(
        model2, 
        X_train_transformed2, # <-- è¿™æ˜¯ä¸€ä¸ªâ€œå¤§â€æ•°æ®é›†
        feature_perturbation="tree_path_dependent"
    )
except Exception as e:
    print(f"è­¦å‘Š: 'tree_path_dependent' å¤±è´¥: {e}")
    print("--- æ­£åœ¨å°è¯• 'interventional' ä½œä¸ºå¤‡é€‰æ–¹æ¡ˆ ---")
    try:
        explainer2 = shap.TreeExplainer(
            model2, 
            X_train_transformed2, 
            feature_perturbation="interventional"
        )
    except Exception as e2:
        print(f"è‡´å‘½é”™è¯¯: æ— æ³•åˆå§‹åŒ– Explainer: {e2}")
        # åœ¨è¿™é‡Œåœæ­¢æˆ–å¼•å‘é”™è¯¯

# --- 
# 4. âœ… è®¡ç®— *ç¼“æ…¢* çš„ SHAP äº¤äº’å€¼ (3D)
# ---
try: 
    combined_interaction_values = explainer2.shap_interaction_values(X_val_transformed2)
    combined_X_val = X_val_transformed2 # (æˆ‘ä»¬ç°åœ¨åªæœ‰ä¸€ä¸ª X_val)
    
    # ----------------------------------------------------
    # 5. âœ… æœ€ç»ˆåˆ†æï¼šè®¡ç®—å¹¶æ‰“å° Top 20 äº¤äº’ç»„åˆ
    # ----------------------------------------------------
    print("\n" + "="*50)
    print(f"--- æ­£åœ¨è®¡ç®— {new_model_descirption} çš„ Top 50 å…¨å±€ [äº¤äº’ç»„åˆ] ---")
    print("="*50)
    
    interaction_matrix = np.abs(combined_interaction_values).mean(axis=0)
    np.fill_diagonal(interaction_matrix, 0) # ç§»é™¤è‡ªæˆ‘äº¤äº’

    features = combined_X_val.columns
    n_features = len(features)

    interaction_pairs = []
    for i in range(n_features):
        for j in range(i + 1, n_features): 
            interaction_strength = interaction_matrix[i, j]
            interaction_pairs.append({
                "feature_1": features[i],
                "feature_2": features[j],
                "strength": interaction_strength
            })

    df_interactions = pd.DataFrame(interaction_pairs)
    df_top_50_interactions = df_interactions.sort_values(
        by="strength", ascending=False
    ).head(50).reset_index(drop=True)

    print("\n--- æ¨¡å‹çš„ Top 50 æœ€ä½³äº¤äº’ç»„åˆ (å…¨å±€) ---")
    print(df_top_50_interactions)

# ---
# âœ… ä¿®æ­£ 1: æ·»åŠ è¿™ä¸ªç¼ºå¤±çš„ 'except' å—
# ---
except Exception as e:
    print(f"\n--- äº¤äº’ç»„åˆ (Interactions) åˆ†æå¤±è´¥ ---")
    print(f"è‡´å‘½é”™è¯¯: {e}")
    # æˆ‘ä»¬å¯ä»¥æ£€æŸ¥ explainer2 æ˜¯å¦è¢«æˆåŠŸåˆ›å»º
    if 'explainer2' not in locals():
        print("é”™è¯¯è¯¦æƒ…: Explainer å¯¹è±¡ 'explainer2' æœªè¢«æˆåŠŸåˆå§‹åŒ–ã€‚")
    else:
        print(f"Explainer ç±»å‹: {type(explainer2)}")
```
